این معیار را بین هر دو هیستوگرام موجود به صورت دستی محاسبه کرده و نتایج را در یک جدول ۱۶ * ۱۶ ارائه دهید. برداشت خود از نتایج را توضیح دهید

همانطور که میدانیم، chi-square distance معیاری است برای فهم میزان تشابه دو مقدار از داده.
به نحوی که هرچقدر این دو دسته از داده که با این معیار مقایسه میشوند، به یکدیگر شبیه تر باشند نتیجه به 0
نزدیکتر بوده و هرچقدر متفاوت باشند نتیجه از صفر به سمت بینهایت دور میشود.

همانطور که واضح است قطر اصلی جدول بالا که بررسی شباهت هر هیستوگرام با خودش است، همگی برابر صفر میباشند که کاملا مورد انتظار است .

همینطور سایر خانه های جدول نیز اعدادی هستند که با بررسی آن ها میتوان میزان شباهت توزیع امتیازات در دو فصل مختلف را بدست آورد. واضحا این اعداد 
هرچقدر کوچکتر باشند یعنی توزیع ها مشابه تر بوده و هرچقدر بزرگتر باشند یعنی توزیع ها بسیار متفاوت بوده اند.

<div dir=rtl>

دقت کنید که ممکن است ماکزیمم( یا مینیمم) امتیازها در یک سال برای دو تیم برابر باشد. در این حالت تیم با تفاضل گل بهتر (یا بدتر) به عنوان خروجی تابع
در نظر گرفته میشود
</div>

```python
def min_points(df_laliga,season):
    df_season = df_laliga[laliga_data['season'] == season][['club', 'points']]
    min_index = df_season['points'].idxmin()
    min_point = df_season.at[min_index, 'points']
    min_club  = df_season.at[min_index, 'club']
    dic = {"club":min_club, "points":min_point}
    return dic

print(min_points(laliga_data, '2006-07'))
```

```python
def max_points(df_laliga,season):
    df_season = df_laliga[laliga_data['season'] == season][['club', 'points']]
    max_index = df_season['points'].idxmax()
    max_point = df_season.at[max_index, 'points']
    max_club  = df_season.at[max_index, 'club']
    dic = {"club":max_club, "points":max_point}
    return dic

print(max_points(laliga_data, '1997-98'))
```

حال از شما می‌خواهیم تابعی بنویسید تا با گرفتن نام فصل جدول آن فصل را خروجی دهد.

 دقت کنید تیم‌ها در جدول یک فصل باید به ترتیب امتیاز و سپس تفاضل گل مرتب شده باشند.

 دقت کنید`DataFrame` لالیگا به عنوان آرگومان اول به تابع ورودی داده می‌شود.
 </div>

```python
def get_season_table(df_laliga,season_name):
    df_season = df_laliga[laliga_data['season'] == season_name]
    df_season = df_season.sort_values(by='goal_difference')
    df_season = df_season.sort_values(by='points')
    return df_season
```

```python
pd.DataFrame.from_dict({'club': {'1970-71': 16,
  '1971-72': 18,
  '1972-73': 18,
  '1973-74': 18,
  '1974-75': 18}})

seasons = []
for i in range(47): seasons.append(f"{1970+i}-{(1970+i+1)%100:02d}")
# print(seasons)
data_cunt = []
for i in range(47): data_cunt.append(len(get_season_table(laliga_data, seasons[i])))
# print(data_c)

club_count = pd.DataFrame(data=data_cunt , columns=['club'], index=seasons)
club_count.index.name = 'season'

club_count.to_csv('club_count.csv', index=True)

```

 دقت کنید که جدولی که ذخیره می‌کنید قالب یکسانی با جدول نمونه ارائه شده در بالا داشته باشد. یعنی `season` همان `index` باشد

سپس بر اساس جدول تکمیل شده در بخش قبلی، تعداد قهرمانی‌های هر تیم را بدست آورید و در متغیر `championship_count` قرار دهید.

```python
all_champions = list(set(data_champion))

dic = {}
for item in all_champions:
    dic[item] = data_champion.count(item)

championship_count = pd.DataFrame(data=list(dic.values()), columns=['club'])
championship_count.index=list(dic.keys())
```

سلول زیر را اجرا کنید تا فایل های csv  را به همراه این نوت‌بوک زیپ کند.

```python
import zipfile
import os


def compress(file_names):
    print("File Paths:")
    print(file_names)

    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED

    # create the zip file first parameter path/name, second mode
    with zipfile.ZipFile("PythonHW1_SaeedJazaeri_98104885.zip", mode="w") as zf:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write('./' + file_name, file_name, compress_type=compression)


file_names = [
    file for file in os.listdir()
    if (file.endswith('csv') or file.endswith('ipynb'))
    and file != 'LaLiga_dataset.csv'
]

compress(file_names)
```
