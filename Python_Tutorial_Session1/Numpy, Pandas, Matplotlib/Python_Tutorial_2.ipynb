# File

### Text file

```python
# read from a file -> readlines():
with open('python_text.txt') as f:
    file_list = f.readlines()

for item in file_list:
    print(item)
```

```python
# remove the ENTER:
for item in file_list:
    print(item.replace('\n', ''))
```

```python
# read from a file -> readline():

with open('python_text.txt') as f:
    item = f.readline()
    while (item):
        print(item.replace('\n', ''))
        item = f.readline()
```

```python
# read from a file -> read():

with open('python_text.txt') as f:
    print(f.read())
```

```python
# "x" - Create - will create a file, returns an error if the file exist

with open('creat_file.txt', 'x') as f:
    f.write('New Line')
```

```python
# "a" - Append - will create a file if the specified file does not exist

with open('creat_file.txt', 'a') as f:
    f.write('\nAppend this line')
```

```python
# "w" - Write - will create a file if the specified file does not exist
#If the file exists it will overwrite the previous file

with open('creat_file.txt', 'w') as f:
    f.write('Write this line\n')
```

# matplotlib

```python
from matplotlib import pyplot as plt
```

```python
x = range(10)
y = range(0,20,2)
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.hist(y, bins=5, rwidth=0.2)
plt.show()
```

```python
x=np.arange(-np.pi,np.pi,0.1)
y1=np.sin(x)
y2=np.cos(x)
```

```python
plt.figure(num=0, figsize=(10,2))
plt.plot(x, y1, 'g--')
plt.plot(x, y2, 'r')
plt.title("this is the title of the plot")
plt.legend(['sinx','cosx'])
plt.ylabel("y axis label")
plt.xlabel("x axis label")
plt.axis([-2*np.pi,2*np.pi,-2,2]) #[xmin xmax ymin ymax]
plt.show()
```

#### Displaying multiple plots using subplot() & subplots()

```python
plt.figure(num=2, figsize=(10,2))
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.sin(x**2)
y3 = y1**2
y4 = y2**2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

ax[0, 0].plot(x, y1, c='red')
ax[0, 1].plot(x, y2, c='red')
ax[1, 0].plot(x, y3, c='blue')
ax[1, 1].plot(x, y3, c='blue')

ax[0, 0].set_title('Simple plot with sin(x)')
ax[0, 1].set_title('Simple plot with sin(x**2)')
ax[1, 0].set_title('Simple plot with sin(x)**2')
ax[1, 1].set_title('Simple plot with sin(x**2)**2')

fig.suptitle('Stacked subplots in two direction')
plt.show()
```

# Class

```python
class student_class():

    def __init__(self, name, age, height):
        self.name = name
        self.age = age
        self.height = height

    def print_name(self):
        print(self.name)

    def print_age(self):
        print(self.age)

    def print_height(self):
        print(self.height)

```

```python
student  = student_class('Ali', 22, 180)
student2 = student_class('hamid', 25, 180)

student2.print_name()
student.print_age()
student.print_height()
```
